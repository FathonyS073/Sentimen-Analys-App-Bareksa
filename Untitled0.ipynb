```python
import pandas as pd

# Membaca dataset
file_path = '/content/drive/MyDrive/Skripsi/Program Skripsi/Hasil/labellingfix.csv'  # Ganti sesuai lokasi file Anda
data = pd.read_csv(file_path)

# Menyesuaikan nama kolom jika diperlukan
data.columns = data.columns.str.strip()  # Menghapus spasi tambahan jika ada

# Fungsi untuk menemukan 5 ulasan tertinggi dan terendah
def find_top_bottom_reviews(df, sentiment, n=5):
    filtered_df = df[df["Sentimen"].str.lower() == sentiment.lower()]
    top_reviews = filtered_df.nlargest(n, "compound scores")
    bottom_reviews = filtered_df.nsmallest(n, "compound scores")
    return top_reviews, bottom_reviews

# Mencari ulasan untuk kelas positif
positive_top, positive_bottom = find_top_bottom_reviews(data, "Positif")
print("Top 5 Positive Sentiment Reviews:")
print(positive_top[["text asli", "compound scores"]])
print("\nBottom 5 Positive Sentiment Reviews:")
print(positive_bottom[["text asli", "compound scores"]])

# Mencari ulasan untuk kelas negatif
negative_top, negative_bottom = find_top_bottom_reviews(data, "Negatif")
print("\nTop 5 Negative Sentiment Reviews:")
print(negative_top[["text asli", "compound scores"]])
print("\nBottom 5 Negative Sentiment Reviews:")
print(negative_bottom[["text asli", "compound scores"]])
```
