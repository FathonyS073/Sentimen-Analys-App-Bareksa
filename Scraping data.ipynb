```python
from google_play_scraper import Sort, reviews, reviews_all
import pandas as pd
import time

# Mengganti 'com.example.app' dengan ID aplikasi yang ingin Anda ambil ulasannya
app_id = 'com.bareksa.app'
num_reviews = 7000  # Jumlah ulasan yang ingin diambil

# Fungsi untuk mengumpulkan ulasan hingga jumlah tertentu
def collect_reviews(app_id, num_reviews):
    all_reviews = []
    batch_size = 200  # Jumlah ulasan yang akan diambil dalam satu panggilan API
    start_idx = 0

    while len(all_reviews) < num_reviews:
        try:
            batch_reviews, _ = reviews(
                app_id,
                lang='id',  # Ganti dengan kode bahasa yang diinginkan
                country='id',  # Ganti dengan kode negara yang diinginkan
                sort=Sort.MOST_RELEVANT,
                count=min(batch_size, num_reviews - len(all_reviews)),
                filter_score_with=None if len(all_reviews) == 0 else 3,  # Hapus untuk mendapatkan semua skor
            )
            all_reviews.extend(batch_reviews)
            start_idx += batch_size

            # Optional: Add a delay to avoid hitting the server too frequently
            time.sleep(1)

        except Exception as e:
            print(f"Error occurred: {e}")
            break

    return all_reviews[:num_reviews]



# Mengumpulkan ulasan
app_reviews = collect_reviews(app_id, num_reviews)

# Konversi ke DataFrame Pandas
df = pd.DataFrame(app_reviews)

```
