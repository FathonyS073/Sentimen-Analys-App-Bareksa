```python
import pandas as pd
import matplotlib.pyplot as plt

# Membaca dataset
data = pd.read_csv('/content/drive/MyDrive/Skripsi/Program Skripsi/Hasil/labellingfix.csv')

# Pastikan kolom 'Sentimen' digunakan untuk perbandingan
label_counts = data['Sentimen'].value_counts()

# Menampilkan perbandingan jumlah label
print("Jumlah masing-masing label:")
print(label_counts)

# Membuat visualisasi perbandingan label
plt.figure(figsize=(8, 6))
label_counts.plot(kind='bar', color=['skyblue', 'salmon'])
plt.title('Perbandingan Label Positif dan Negatif', fontsize=14)
plt.xlabel('Sentimen', fontsize=12)
plt.ylabel('Jumlah', fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Memisahkan data positif dan negatif
data_positif = data[data['Sentimen'] == 'Positif']
data_negatif = data[data['Sentimen'] == 'Negatif']

# Mengambil 5 nilai tertinggi untuk label positif
top_positif = data_positif.nlargest(5, 'compound scores')

# Mengambil 5 nilai terendah untuk label negatif
top_negatif = data_negatif.nsmallest(5, 'compound scores')

# Menampilkan hasil
print("5 Positif Tertinggi:")
print(top_positif[['text asli', 'compound scores', 'Sentimen']])

print("\n5 Negatif Tertinggi:")
print(top_negatif[['text asli', 'compound scores', 'Sentimen']])

# Membuat visualisasi 5 positif tertinggi
plt.figure(figsize=(10, 6))
plt.barh(top_positif['text asli'], top_positif['compound scores'], color='green')
plt.title('5 Positif Tertinggi berdasarkan Compound Scores', fontsize=14)
plt.xlabel('Compound Scores', fontsize=12)
plt.ylabel('Text Asli', fontsize=12)
plt.gca().invert_yaxis()
plt.show()

# Membuat visualisasi 5 negatif tertinggi
plt.figure(figsize=(10, 6))
plt.barh(top_negatif['text asli'], top_negatif['compound scores'], color='red')
plt.title('5 Negatif Tertinggi berdasarkan Compound Scores', fontsize=14)
plt.xlabel('Compound Scores', fontsize=12)
plt.ylabel('Text Asli', fontsize=12)
plt.gca().invert_yaxis()
plt.show()
```

```python
import pandas as pd
from collections import Counter
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Membaca dataset
data = pd.read_csv('/content/drive/MyDrive/Skripsi/Program Skripsi/Hasil/labellingfix.csv')

# Memisahkan data positif dan negatif
data_positif = data[data['Sentimen'] == 'Positif']
data_negatif = data[data['Sentimen'] == 'Negatif']

# Menggabungkan teks untuk masing-masing sentimen
text_positif = ' '.join(data_positif['text_prepro'].astype(str))
text_negatif = ' '.join(data_negatif['text_prepro'].astype(str))

# Tokenisasi dan menghitung frekuensi kata
frekuensi_positif = Counter(text_positif.split())
frekuensi_negatif = Counter(text_negatif.split())

# Menampilkan 10 kata paling sering muncul di masing-masing label
print("10 Kata Paling Sering Muncul di Label Positif:")
print(frekuensi_positif.most_common(10))

print("\n10 Kata Paling Sering Muncul di Label Negatif:")
print(frekuensi_negatif.most_common(10))

# Visualisasi WordCloud untuk kata-kata paling sering muncul
def buat_wordcloud(frekuensi, title, colormap):
    wordcloud = WordCloud(width=800, height=400, background_color='white', colormap=colormap).generate_from_frequencies(frekuensi)
    plt.figure(figsize=(10, 6))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(title, fontsize=14)
    plt.axis('off')
    plt.show()

buat_wordcloud(frekuensi_positif, 'WordCloud Kata Terpopuler - Positif', 'Greens')
buat_wordcloud(frekuensi_negatif, 'WordCloud Kata Terpopuler - Negatif', 'Reds')
```
